PREPROCESSING

```python
import pandas as pd
data = pd.read_csv("cosmetic.csv")
```

```python
data
```

```python
data.info()
```

```python
data = data.loc[pd.notnull(data['ingredients'])]
```

```python
data.isnull().sum()
```

```python
data.Label[data['Label'] == 'moisturizing-cream-oils-mists'] = str('Moisturizer')
data.Label[data['Label'] == 'cleanser'] = str('Cleanser')
data.Label[data['Label'] == 'facial-treatments'] = str('Treatment')
data.Label[data['Label'] == 'face-mask'] = str('Face Mask')
data.Label[data['Label'] == 'eye-treatment-dark-circle-treatment'] = str('Eye cream')
data.Label[data['Label'] == 'sunscreen-sun-protection'] = str('Sun protect')
```

```python
data
```

```python
data2 = data['name'].drop_duplicates()
data = data.loc[data2.index, :].reset_index()
```

```python
data.drop(['index'], axis = 1, inplace = True)
```

```python
data
```

```python
import re
pattern = re.compile(r"(\d+).\d+")
for i in range(len(data)):
    data['price'][i] = re.findall(pattern, data['price'][i])[0]
```

```python
data['price'] = pd.to_numeric(data['price'])
```

```python
data['rank'].fillna(0, inplace = True)
```

```python
pattern = re.compile(r"([a-zA-Z]+)\\n")
for i in range(len(data)):
    data['skin_type'][i] = re.findall(pattern, data['skin_type'][i])
```

```python
data
```

```python
df_2 = data['skin_type'].str.join('|').str.get_dummies()
data2 = data.join(df_2).drop('skin_type', axis = 1)
```

```python
data['skin_type']
```

```python
a = [t.split('\r\n\r\n') for t in data['ingredients']]
pattern = ['\r\n', '-\w+: ', 'Please', 'No Info', 'This product', 'Visit']
for i in range(len(data)):
    Num = len(a[i])
    for j in range(Num):
        if all(x not in a[i][j] for x in pattern):
           data2['ingredients'][i] = a[i][j]
```

```python
data2.to_csv('cosmetic_prep.csv', encoding = 'utf-8-sig', index = False)
```
